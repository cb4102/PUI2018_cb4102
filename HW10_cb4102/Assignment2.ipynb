```python
#establish a variable for the predicted model:
#endogenous variable is the y variable

#ols_quad.compare_lr_test(ols1)

pred = ols_quad2.predict(building_energy_clean_df)
```

```python
def chi_squared(data, model, errors=False):
    '''Calculates the chi sq given data, model and errors
    Arguments:
    data: series of datapoints (endogenous variable)
    model: series of predicted values corresponding to the observed data
    errors: serie of errors (optional). 
    If errors are not passes all errors are set to 1
    ''' 
    if errors is False:
        return chi2(data,model)
    else:
        data_errors = data + np.sqrt(data)
        model_errors = model + np.sqrt(model)
        
        return chi2(data_errors,model_errors)
```

```python
#calculate errors for units and energy
#error(log10(x)) = (sqrt(x) / (x / log(10))

building_energy_clean_df['energy_errors'] = np.sqrt(building_energy_clean_df['total_energy'])
building_energy_clean_df['units_errors'] = np.sqrt(building_energy_clean_df['UnitsTotal'])
building_energy_clean_df['total_error'] = np.sqrt(building_energy_clean_df['energy_errors'] ** 2 \
                                                  + building_energy_clean_df['units_errors'] ** 2)

building_energy_clean_df['energylog_errors'] = np.sqrt(building_energy_clean_df['total_energy']) / \
                                                    (building_energy_clean_df['total_energy'] * np.log(10))
    

building_energy_clean_df['unitslog_errors'] = np.sqrt(building_energy_clean_df['UnitsTotal']) / \
                                                    (building_energy_clean_df['UnitsTotal'] * np.log(10))
    
building_energy_clean_df['totallog_error'] = np.sqrt(building_energy_clean_df['energylog_errors'] ** 2 \
                                                     + building_energy_clean_df['unitslog_errors'] ** 2)
```
